# Quadratic integrate-and-fire neurons

The QIF model describes the neuron's voltage dynamics through the ODE

$$ \frac{dV(t)}{dt} = V(t)^2 + b,  \text{ if } V > V_{th} \text{ then } V=V_{reset}, $$

with $b$ a constant (e.g. input current).
